```python
import os 
def get_track(path):
    track_data = pd.DataFrame()
    data = json.load(open(path, encoding='utf-8'))
    df = pd.DataFrame(data["playlists"])
    for i in range(len(df)):
        df_tracks = df.loc[df['name'] == df['name'][i], 'tracks'].values[0]
        df_tracks = pd.DataFrame(df_tracks)
        df_tracks.drop('pos', axis=1, inplace=True)
        track_data = pd.concat([track_data, df_tracks], axis=0)
    return track_data


def get_all_track(path):
    files = os.listdir(path)
    all_tracks = pd.DataFrame()
    for i in files:
        new_path = path + '//'+ i
        try:
            track_data = get_track(new_path)
            all_tracks = pd.concat([all_tracks, track_data], axis=0)
        except IndexError:
            print(f"No tracks found in file: {i}")
    all_tracks.drop_duplicates(inplace=True)   
    return all_tracks
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# One-hot encode the artist names
preprocessor = ColumnTransformer(
	transformers=[
		('artist_name', OneHotEncoder(), ['artist_name'])
	])

cluster_pipeline = Pipeline([('preprocessor', preprocessor), ('scaler', StandardScaler(with_mean=False)), ('kmeans', KMeans(n_clusters=10))])
X = all_tracks[['artist_name']]
cluster_pipeline.fit(X)
all_tracks['cluster'] = cluster_pipeline.predict(X)
```
