```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'..\data\processed\sample_dataset\tracks_modified.csv')
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder

# Use LabelEncoder to convert track names to numeric labels
label_encoder = LabelEncoder()
df['track_id'] = label_encoder.fit_transform(df['track_name'].astype(str))

df.head(5)
```

```python
import scipy.sparse as sp
from implicit.als import AlternatingLeastSquares
from implicit.evaluation import precision_at_k

playlist_song_matrix = sp.csr_matrix(
    (np.ones(len(df)), (df["pid"], df["track_id"]))
)


model = AlternatingLeastSquares(factors=50, regularization=0.1, iterations=20)
model.fit(playlist_song_matrix.T)
```
